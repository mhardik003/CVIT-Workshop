### Problem 1

Given a list of binary strings `numbers`. Define a function that takes the first `k` strings and concats their parity into a string `encoding` and returns it.

````python
#@title
def func1 (numbers, k):
  encoding = ""
  for index in range(k):
    parity= 0
    for digit in numbers[index]:
      parity = parity^int(digit)
    
    encoding = encoding + str(parity)

  return encoding


print(func1(["000","000"], 2))
print(func1(["111","111"], 2))
print(func1(["101","010"], 2))
````

**Solution**:

````python
def get_encoding(numbers,k):
  encoding = ""
  for index in range(k):
    parity = 0
    for digit in numbers[index]:
        parity ^= int(digit)
    encoding += str(parity)
  return encoding
````

**Do a simple thresholding on the image sudoku.jpg, and find an apt threshold for the same using the above method.** 

````python
sudoku = cv2.imread("./images/sudoku.jpg", cv2.IMREAD_GRAYSCALE)


fig, axs = plt.subplots(2, 2)

ret, sudoku_binthresh = cv2.threshold(sudoku, 85, 255, cv2.THRESH_BINARY)
sudoku_adaptivethresh_gaussian = cv2.adaptiveThreshold(
    sudoku, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 7
)
sudoku_adaptivethresh_mean = cv2.adaptiveThreshold(
    sudoku, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 10
)

axs[0][0].imshow(sudoku, cmap="gray")
axs[0][0].set_title("Original Image")

axs[0][1].imshow(sudoku_binthresh, cmap="gray")
axs[0][1].set_title("Binary thresholding")

axs[1][0].imshow(sudoku_adaptivethresh_gaussian, cmap="gray")
axs[1][0].set_title("Adaptive gaussian thresholding")

axs[1][1].imshow(sudoku_adaptivethresh_mean, cmap="gray")
axs[1][1].set_title("Adaptive mean thresholding")

fig.tight_layout()
````
